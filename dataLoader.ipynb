```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as mpimg


import random

```

```python
images = np.load('images.npy')
cats = np.load('cats.npy')

images.shape
```

```python
numImages = len(images)
```

```python
# generate 10 random numbers between 0 and the number of images
randomList = random.sample(range(0, numImages), 30)
```

```python
for i in randomList:
    
    plt.figure()
    imgplot = plt.imshow(images[i])
    plt.axis('off')
    plt.title(cats[i])
```
